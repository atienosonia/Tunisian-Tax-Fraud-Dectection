## Data Cleaning

Duplicates 

```python
duplicates = []

def check_duplicates(data):

    """A simple function to check the duplicates in the dataset"""

    for i in data.duplicated():
        duplicates.append(i)
    duplicates_set = set(duplicates)
    if (len(duplicates_set) == 1):
        print('The Dataset has no duplicates')

    else:
        duplicates_percentage = np.round(((sum(duplicates)/len(data)) * 100 ), 2)
        print(f'Duplicated rows constitue of {duplicates_percentage} of the dataset')

check_duplicates(df)
```

Missing Values

```python
def missing_values(data):

    """A simple function that returns a dataframe of the missing values in the dataste"""

    missing_values = data.isnull().sum().sort_values(ascending=False)

    missing_val_percent = ((data.isnull().sum()/len(data)).sort_values(ascending=False))

    """creating a dataframe containing missing values and their percentages"""

    missing_df = pd.DataFrame({'Missing Vlaues' : missing_values, 'Percentage %' : missing_val_percent})

    return missing_df[missing_df['Percentage %'] > 0.6]

missing_values(df)

```
